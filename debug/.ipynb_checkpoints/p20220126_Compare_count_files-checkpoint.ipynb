```python
df_eth['pipeline'] = 'graph-based'

df_ohsu = df_ohsu.loc[df_ohsu['normal_cohort_id'] == 'core_GTEx', :]
df_eth = df_eth.loc[df_eth['normal_cohort_id'] == 'Gtexcore', :]

df_ohsu['normal_cohort_id'] = 'Gtexcore'

col_interest = df_ohsu.columns
df_ohsu.columns = [col.replace('Background', 'Normal') for col in col_interest]
```

```python
for col in merge_: 
    print(col)
    print(df_ohsu[col].dtypes)
    print(df_eth[col].dtypes)
    if (df_ohsu[col].dtypes != df_eth[col].dtypes) and col != 'pipeline':
        print('change type')
        df_ohsu[col] = df_ohsu[col].astype( df_eth[col].dtypes ) 
    print('\n')
```

```python
df_m = df_eth.merge(df_ohsu, on = merge_, how = 'outer')
```

```python
save = False
tag = ''
```

```python
def plot_stacked_barplot(Y, front_plot, sorting_list, ascending=True):
    from matplotlib.lines import Line2D
    plt.figure(figsize = (3, 5))
    df = df_m.sort_values(sorting_list, ascending=ascending)
    color_ohsu = 'lightsalmon'
    color_eth = 'steelblue'
    if front_plot == 'ohsu':
        sns.barplot(data = df.loc[df['pipeline'] == 'junction-based', :],
                    x = 'experiment_id', y= Y, color = color_ohsu)
        sns.barplot(data = df.loc[df['pipeline'] == 'graph-based', :],
                    x = 'experiment_id', y= Y, color = color_eth)
    else:
        sns.barplot(data = df.loc[df['pipeline'] == 'graph-based', :],
                    x = 'experiment_id', y= Y, color = color_eth)
        sns.barplot(data = df.loc[df['pipeline'] == 'junction-based', :],
                            x = 'experiment_id', y= Y, color = color_ohsu)
    custom_lines = [Line2D([0], [0], color=color_ohsu, lw=4),\
    Line2D([0], [0], color=color_eth, lw=4)]
    plt.legend(custom_lines, ["% kmer after filtering step relative to initial OHSU generated kmers",\
               '% kmer after filtering step relative to initial ETH generated kmers'])
    plt.xticks(rotation=90)
    if save:
        print("save to {}".format(save_path))
        plt.savefig(save_path, bbox_inches='tight')
```

```python
def plot_side_barplot(Y, front_plot, sorting_list, ascending=True):
    from matplotlib.lines import Line2D
    plt.figure(figsize = (3, 5))
    df = df_m.sort_values(sorting_list, ascending=ascending)
    palette ={"junction-based": "lightsalmon", "graph-based": "steelblue"}
    sns.barplot(data = df,
                    x = 'experiment_id', y= Y, hue = 'pipeline', palette = palette)
    custom_lines = [Line2D([0], [0], color=palette["junction-based"], lw=4),\
    Line2D([0], [0], color=palette["graph-based"], lw=4)]
    plt.legend(custom_lines, ["% kmer after filtering step relative to initial OHSU generated kmers",\
               '% kmer after filtering step relative to initial ETH generated kmers'])
    plt.xticks(rotation=90)
    save_path = os.path.join(plot_dir, '{}_{}_{}_{}_SeqProportionPlot.pdf'.format(base_plot, sample, Y, tag ))  

    if save:
        print("save to {}".format(save_path))
        plt.savefig(save_path, bbox_inches='tight')
```
