```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
import tarfile
import pathlib
import glob
import pathlib
```

```python
# Inputs ETH
run_type = 'ov'

if run_type == 'brca': # Note in this case the filtering ids are the same but could differ in case of rerun
    basedir = '/cluster/work/grlab/projects/projects2020_OHSU/peptides_generation/CANCER_eth/commit_c4dd02c_conf2_Frame_cap0_runs/TCGA_Breast_1102'
    filtering_id_STAR = 'filters_19May_order_5ge_wAnnot_STAR'
    filtering_id_GP = 'filters_19May_order_5ge_wAnnot'
    filtering_id_inter = 'filters_19May_order_5ge_wAnnot_GPstar'
elif run_type == 'ov':
    basedir = '/cluster/work/grlab/projects/projects2020_OHSU/peptides_generation/CANCER_eth/commit_c4dd02c_conf2_Frame_cap0_runs/TCGA_Ovarian_374'
    filtering_id_STAR = 'filters_19May_order_5ge_wAnnot_STAR'
    filtering_id_GP =  'filters_19May_order_5ge_wAnnot'
    filtering_id_inter = 'filters_19May_order_5ge_wAnnot_GPstar'
```

```python
# Prepare intersection path
outpath = os.path.join(basedir, 'filtering_samples', filtering_id_inter)
pathlib.Path(outpath).mkdir(exist_ok = True)
```

```python
# Filtering names 
pathS_filt_eth = glob.glob(os.path.join(basedir, 'filtering_samples', filtering_id_GP, '*Filt*tsv.gz'))
print(len(pathS_filt_eth))
pathS_filt_eth = [os.path.basename(i) for i in pathS_filt_eth]
```

# Load

```python
for path_filt_eth in pathS_filt_eth: 
    path_filt_eth_STAR = os.path.join(basedir, 'filtering_samples', filtering_id_STAR, path_filt_eth)
    path_filt_eth_GP = os.path.join(basedir, 'filtering_samples', filtering_id_GP, path_filt_eth)

    # Read ETH Data
    print(f'Reading {path_filt_eth_GP}')
    df_filt_GP = pd.read_csv(path_filt_eth_GP, sep = '\t')
    print(df_filt_GP.shape)

    print(f'Reading {path_filt_eth_STAR}')
    df_filt_STAR = pd.read_csv(path_filt_eth_STAR, sep = '\t')
    print(df_filt_STAR.shape)




    # Some intersection Numbers
    F_GP = set(df_filt_GP['kmer'])
    print(len(F_GP))
    F_STAR = set(df_filt_STAR['kmer'])
    print(len(F_STAR))
    print(len(F_GP.intersection(F_STAR)))

    # Intersection coord Combination rule for filtered datasets
    intersect_STAR_GP = df_filt_GP.merge(df_filt_STAR[['kmer']], on = ['kmer'], how = 'inner')
    unique_kmers = len(set(intersect_STAR_GP['kmer']))
    print(f'Corresponds to {unique_kmers} unique kmers')

    outfile = os.path.join(outpath, path_filt_eth)
    intersect_STAR_GP.to_csv(outfile, 
                             index=None, sep = '\t', compression = 'gzip')
    print(f'Saving to {outfile}')
```
