```python
# read files as list 
kmers_dict = [[],[], []]
with tarfile.open(tar_file_OHSU, "r:*") as tar:
    for io, _ in enumerate(file_pair['eth']):
        #print(io)
        name_ohsu = file_pair['ohsu'][io]
        eth_path = file_pair['eth'][io]
        df_ohsu = pd.read_csv(tar.extractfile(name_ohsu), header=0, sep="\t")
        df_eth = pd.read_csv(eth_path, sep="\t", usecols = ['kmer'])
        df_eth = set(df_eth['kmer'])
        df_ohsu = set(df_ohsu['kmer'])
        for kmer in df_eth:
            kmers_dict[0].append(kmer)
            kmers_dict[1].append(eth_path.split('/')[-2])
            kmers_dict[2].append(io)
        for kmer in df_ohsu:
            kmers_dict[0].append(kmer)
            kmers_dict[1].append(name_ohsu)
            kmers_dict[2].append(io)
```

```python
# create summary dataframe
kmers_df = pd.DataFrame(np.array(kmers_dict).T, columns = ['kmer', 'filename', 'pair_id'])
```

```python
# create columns of interest 
kmers_df['sample'] = [os.path.basename(name).split('_')[1] for name in kmers_df['filename']]

kmers_df['pipeline'] = [os.path.basename(name).split('_')[0] for name in kmers_df['filename']]

kmers_df['experiment'] =  ['_'.join(os.path.basename(name).split('_')[2:]) for name in kmers_df['filename']]



kmers_df = kmers_df.drop(['filename'], axis = 1)
```

```python
kmers_df.tail()
```

```python
# label shared kmers 
shared_kmers =   kmers_df.loc[:, ['kmer', 'pair_id', 'experiment']].groupby(['pair_id', 'kmer']).count().reset_index()
shared_kmers = shared_kmers.rename({'experiment': 'shared'}, axis = 1)

shared_kmers.head()

kmers_df.head()

kmers_df.shape

kmers_df_with_s = kmers_df.merge(shared_kmers, on = ['kmer', 'pair_id'], how = 'left')

kmers_df_with_s.head()
```

```python
# Count: Normalisation factor (Initial number of kmers before filtering)
res = [[], [], []]
for sample in kmers_df_with_s['sample'].unique():
    # add samples
    res[0].extend([sample]*2 )
    sample = sample.split('.')[0]
    
    # add ohsu initial number of kmer in sample
    name_ohsu = f'{cancer_cohort_ohsu}/J_filtered_df_{sample}_samp_chrt_norm_mot.tsv'
    with tarfile.open(tar_counts_OHSU, "r:*") as tar:
        df_ohsu = pd.read_csv(tar.extractfile(name_ohsu), header=0, sep="\t")
        res[1].append('J')
        res[2].append(df_ohsu['Init_cancer'].unique()[0])
        
    # add eth initial number of kmer in sample    
    path_eth = os.path.join(base_folder_ETH, 
               f'filter_{sample}.all/{commit_folder_ETH[:-1]}_counts/G_filtered_df_{sample}_samp_chrt_norm_mot_unip.tsv')
    df_eth = pd.read_csv(path_eth, sep = '\t')
    res[1].append('G')
    res[2].append(df_eth.loc[df_eth['normal_cohort_id'] == 'Gtexcore' , 
                 'Init_cancer'].unique()[0]) # # TODO: for now hard coded, 
                 #in the future transform to metho for all the backgrounds (batch)

kmers_df_grp2 = pd.DataFrame(res).T
kmers_df_grp2.columns = ['sample', 'pipeline', 'init_kmer_per_experiment']
```

```python
# Compute percentages 
kmers_df_grp_counts['kmer_per_experiment_%of_of_most_lenient'] = 0 
kmers_df_grp_counts['shared_kmer_per_experiment_%of_of_most_lenient'] = 0 
kmers_df_grp_counts['kmers_init_in_sample'] = 0 
round_digits = 4
for idx, row in kmers_df_grp2.iterrows():
    sample, pipeline, tot_most_lenient  = row['sample'], row['pipeline'], row['init_kmer_per_experiment']
    
    subset = kmers_df_grp_counts.loc[(kmers_df_grp_counts['sample'] ==  sample) &  \
                           ( kmers_df_grp_counts['pipeline'] == pipeline ), :].index
    
    kmers_df_grp_counts.loc[ subset ,'kmer_per_experiment_%of_of_most_lenient'] = \
    np.round(kmers_df_grp_counts.loc[ subset ,'kmer_per_experiment'] / tot_most_lenient * 100 , round_digits)

    kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment_%of_of_most_lenient'] = \
    np.round(kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment'] / tot_most_lenient * 100, round_digits)
    
#     kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment_%of_of_most_lenient'] = \
#     np.round(kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment'] / tot_most_lenient * 100 , round_digits)
    
    kmers_df_grp_counts.loc[ subset ,'kmers_init_in_sample'] = tot_most_lenient
   
```

```python
def get_experiment_order(kmers_df_grp_counts, sample, pipeline, pipeline2):
    '''Get experiment order based on sample in pipeline', and get associated names in pipeline 2'''
    names_ordered = {}
    experiment_order = kmers_df_grp_counts.loc[(kmers_df_grp_counts['pipeline'] == pipeline) & \
                    (kmers_df_grp_counts['sample'] == sample), :].\
                    sort_values(by=['kmer_per_experiment'], 
                    ascending = False)[['pair_id', 'experiment']]
    names_ordered[f'names_{pipeline}'] = experiment_order['experiment'].values
    names_ordered[f'names_{pipeline2}'] = [kmers_df_grp_counts.loc[(kmers_df_grp_counts['pipeline'] == pipeline2) \
                        & (kmers_df_grp_counts['sample'] == sample) 
                        & (kmers_df_grp_counts['pair_id'] == pair_id), 'experiment'].values[0]
                        for pair_id in experiment_order['pair_id']]
    return names_ordered

    
```

```python
def get_multiple_exp_order(kmers_df_grp_counts, samples):
    store_order = {}
    for sample in samples:
        store_order[sample] = {}
        store_order[sample]['order_J'] = get_experiment_order(kmers_df_grp_counts, 
                                                              sample, 'J', 'G')
        store_order[sample]['order_G'] = get_experiment_order(kmers_df_grp_counts, 
                                                        sample, 'G', 'J')
    return store_order
        
        
```

```python
def plot_stacked_barplot(plot_df, store_order, apply_order, pipeline):
    '''save, plot_dir, base_plot are global variables'''
    plot_df = plot_df.loc[plot_df['pipeline'] == pipeline]

    cmap = plt.cm.coolwarm
    c2 = 'green'
    if pipeline == 'J':
        pipeline_sharing = 'G'
        c1 = cmap(0.2)
    else:
        pipeline_sharing = 'J'
        c1 = cmap(0.8)
        

    for sample in plot_df['sample'].unique():
        print(sample)
        experiment_order = store_order[sample][f'order_{apply_order}'][f'names_{pipeline}']

        

        plt.figure(figsize = (25, 10))
        sns.barplot(x="experiment", y="kmer_per_experiment_%of_of_most_lenient", \
                    data=plot_df.loc[plot_df['sample'] == sample], color = c1, order = experiment_order)
        sns.barplot(x="experiment", y="shared_kmer_per_experiment_%of_of_most_lenient", \
                    data=plot_df.loc[plot_df['sample'] == sample], color = c2, order = experiment_order)
        plt.xticks(rotation=90)
        plt.ylabel('Proportion kmers retained after filtering (%) sample)', fontsize = 20)
        custom_lines = [Line2D([0], [0], color=c1, lw=4),\
                        Line2D([0], [0], color=c2, lw=4)]
        plt.legend(custom_lines, ["Kmers in {} pipeline".format(pipeline),\
                           "Proportion also found in {} pipeline".format(pipeline_sharing)])
        if save:
            save_path = os.path.join(plot_dir, 
                                    '{}_{}_{}_{}_ProportionPlot.pdf'.format(base_plot, 
                                     sample, pipeline, f'order{apply_order}' ))  
            print("save to {}".format(save_path))
            plt.savefig(save_path, bbox_inches='tight')
        plt.show()
```

```python
def inverted_stack_barplot(plot_df, store_order, apply_order):
    c1 = 'green'
    y_lim = 1.8
    fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,10))

    for ax, pipeline, invert, c2 in zip(axes, ['J', 'G'], [False, True], [cmap(0.2), cmap(0.8)]):
        experiment_order = store_order[sample][f'order_{apply_order}'][f'names_{pipeline}']
        plot_df = kmers_df_grp_counts.loc[kmers_df_grp_counts['pipeline'] == pipeline]
        plot_df = plot_df.loc[plot_df['sample'] == sample].set_index('experiment')
        y = plot_df.loc[experiment_order, "kmer_per_experiment_%of_of_most_lenient"].values
        y_sh = plot_df.loc[experiment_order,"shared_kmer_per_experiment_%of_of_most_lenient"].values
        x = experiment_order
        ax.bar(x, y, color = c2)
        ax.bar(x, y_sh, color = c1)
        ax.set_ylim((0, y_lim))  
        custom_lines = [Line2D([0], [0], color=c2, lw=4),\
                        Line2D([0], [0], color=c1, lw=4)]
        ax.legend(custom_lines, ["Kmers in {} pipeline".format(pipeline),\
                       "Shared across pipelines".format(pipeline_sharing)])
        if invert:
            ax.invert_yaxis()
    plt.suptitle(f'Filtering of sample {sample}')
    plt.ylabel(f'% kmers retained', fontsize = 12)
    plt.xticks(rotation=90)
    plt.subplots_adjust(hspace=0)
    if save:
        save_path = os.path.join(plot_dir, 
                                '{}_{}_{}_ProportionPlotInverted.pdf'.format(base_plot, 
                                 sample, f'order{apply_order}' ))  
        print("save to {}".format(save_path))
        plt.savefig(save_path, bbox_inches='tight')
    plt.show()
```
