```python
kmers_dict = [[],[], []]
with tarfile.open(tar_file_OHSU, "r:*") as tar:
    for io, _ in enumerate(file_pair['eth']):
        #print(io)
        name_ohsu = file_pair['ohsu'][io]
        eth_path = file_pair['eth'][io]
        df_ohsu = pd.read_csv(tar.extractfile(name_ohsu), header=0, sep="\t")
        df_eth = pd.read_csv(eth_path, sep="\t", usecols = ['kmer'])
        df_eth = set(df_eth['kmer'])
        df_ohsu = set(df_ohsu['kmer'])
        for kmer in df_eth:
            kmers_dict[0].append(kmer)
            kmers_dict[1].append(eth_path.split('/')[-2])
            kmers_dict[2].append(io)
        for kmer in df_ohsu:
            kmers_dict[0].append(kmer)
            kmers_dict[1].append(name_ohsu)
            kmers_dict[2].append(io)
```

```python
kmers_df = pd.DataFrame(np.array(kmers_dict).T, columns = ['kmer', 'filename', 'pair_id'])
```

```python
# Compute percentages of most lenient 
kmers_df_grp_counts['kmer_per_experiment_%of_of_most_lenient'] = 0 
kmers_df_grp_counts['shared_kmer_per_experiment_%of_of_most_lenient'] = 0 
kmers_df_grp_counts['kmers_most_lenient_pipelinesample'] = 0 
for idx, row in kmers_df_grp2.iterrows():
    sample, pipeline, tot_most_lenient  = row['sample'], row['pipeline'], row['max_kmer_per_experiment']
    
    subset = kmers_df_grp_counts.loc[(kmers_df_grp_counts['sample'] ==  sample) &  \
                           ( kmers_df_grp_counts['pipeline'] == pipeline ), :].index
    
    kmers_df_grp_counts.loc[ subset ,'kmer_per_experiment_%of_of_most_lenient'] = \
    np.round(kmers_df_grp_counts.loc[ subset ,'kmer_per_experiment'] / tot_most_lenient, 2)

    kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment_%of_of_most_lenient'] = \
    np.round(kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment'] / tot_most_lenient, 2)
    
    kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment_%of_of_most_lenient'] = \
    np.round(kmers_df_grp_counts.loc[ subset ,'shared_kmer_per_experiment'] / tot_most_lenient, 2)
    
    kmers_df_grp_counts.loc[ subset ,'kmers_most_lenient_pipelinesample'] = tot_most_lenient
   
```

```python
def get_experiment_order(kmers_df_grp_counts, pipeline, sample):
    '''Get experiment order for one pipeline and one sample'''
    if pipeline == 'J':
        pipeline2 = 'G'
    else:
        pipeline2 = 'J'
    experiment_order = kmers_df_grp_counts.loc[(kmers_df_grp_counts['pipeline'] == pipeline) & \
                    (kmers_df_grp_counts['sample'] == sample), :].\
        sort_values(by=['kmer_per_experiment'], ascending = False)[['pair_id', 'experiment']]
    experiments_1 = experiment_order['experiment']
    experiments_2 = [kmers_df_grp_counts.loc[(kmers_df_grp_counts['pipeline'] == pipeline2) & \
                        (kmers_df_grp_counts['sample'] == sample) & \
                        (kmers_df_grp_counts['pair_id'] == pair_id), 'experiment'].values[0]
                        for pair_id in experiment_order['pair_id']]
    if pipeline == 'J':
        return experiments_1, experiments_2
    else:           
        return experiments_2, experiments_1

    
```

```python
def plot_stacked_barplot(plot_df, experiment_order, pipeline, pipeline_sharing, tag):
    '''save, plot_dir, base_plot are global variables'''
    for sample in plot_df['sample'].unique():
        save_path = os.path.join(plot_dir, '{}_{}_{}_{}_ProportionPlot.pdf'.format(base_plot, sample, pipeline, tag ))  
        print(sample)
        cmap = plt.cm.coolwarm
        c1 = cmap(0.2)
        c2 = 'orange'
        plt.figure(figsize = (25, 10))
        sns.barplot(x="experiment", y="kmer_per_experiment_%of_of_most_lenient", \
                    data=plot_df.loc[plot_df['sample'] == sample], color = c1, order = experiment_order)
        sns.barplot(x="experiment", y="shared_kmer_per_experiment_%of_of_most_lenient", \
                    data=plot_df.loc[plot_df['sample'] == sample], color = c2, order = experiment_order)
        plt.xticks(rotation=90)
        plt.ylabel('Number of kmers (% of the most lenient Gtexcore filtered experiment)', fontsize = 20)
        custom_lines = [Line2D([0], [0], color=c1, lw=4),\
                Line2D([0], [0], color=c2, lw=4)]
        plt.legend(custom_lines, ["Kmers in {} pipeline".format(pipeline),\
                           "Proportion also found in {} pipeline".format(pipeline_sharing)])
        if save:
            print("save to {}".format(save_path))
            plt.savefig(save_path, bbox_inches='tight')
        plt.show()
```
